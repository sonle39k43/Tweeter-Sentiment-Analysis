```python
import os
import pandas as pd
from datetime import date
import yake
import numpy as np
import re
import nltk
from textblob import TextBlob
```

```python
def cleantwt(twt):
    twt = re.sub('#bitcoin','bitcoin',twt)
    twt = re.sub('#Bitcoin','Bitcoin',twt)
    twt = re.sub('#ethereum','ethereum',twt)
    twt = re.sub('#Ethereum','Ethereum',twt)
    twt = re.sub('#[A-Za-z0-9]+','',twt)
    twt = re.sub('\\n','',twt)
    twt = re.sub('https?:\/\/\S','',twt)
    lis = twt.split()
    res = []
    for char in lis:
        if char[0] != '@':
            res.append(char)
    twt = ' '.join(res)
    return twt
```

```python
language = "en"
max_ngram_size = 2
deduplication_thresold = 0.9
deduplication_algo = 'seqm'
windowSize = 1
numOfKeywords = 5
def extract_keywords(twt):
    custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_thresold, dedupFunc=deduplication_algo, windowsSize=windowSize, top=numOfKeywords, features=None)
    keywords = custom_kw_extractor.extract_keywords(twt)
    res = []
    for i in range(len(keywords)):
        res.append(keywords[i][0])
    return res
```

```python
column = {"Tweet_bitcoin":[]}
with open('extracted-bitcoin.txt', 'r') as in_file:
      for line in in_file:
        column["Tweet_bitcoin"].append(line)
df = pd.DataFrame(column)
```

```python
df.head()
```

```python
df['Cleaned_Tweets'] = df['Tweet_bitcoin'].apply(cleantwt)
df.head()
```

```python
def getSubjectivity(twt):
  return TextBlob(twt).sentiment.subjectivity
def getPolarity(twt):
  return TextBlob(twt).sentiment.polarity 

df['Subjectivity'] = df['Cleaned_Tweets'].apply(getSubjectivity)
df['Polarity'] = df['Cleaned_Tweets'].apply(getPolarity)

df.head()
```

```python
def getSentiment(score):
  if score < 0:
    return 'Negative'
  elif score == 0:
    return 'Neutral'
  else:
    return 'Postive'

df['Sentiment'] = df['Polarity'].apply(getSentiment)
df.head()
```

```python
import math
def get_score24h():
    avg = df['Polarity'].mean()
    if avg>0:
        angle = 90-(math.acos(avg)*180/math.pi) + 90
    else:
        angle = math.acos(-avg)*180/math.pi
    score = angle*10/180
    return score
```

```python
df['score_btc'] = get_score24h()
df.head()
```

```python
df['key_word_btc'] = df['Cleaned_Tweets'].apply(extract_keywords)
df.head()
```

```python
import json
import pandas as pd
# Make it work for Python 2+3 and with Unicode
import io
import re
from datetime import date
import datetime
import nltk
from nltk.stem import PorterStemmer
import nltk.corpus
stemmer = PorterStemmer()
words = set(nltk.corpus.words.words())
today = date.today()
str1= str(today.year)[2:]+'/'+str(today.month)+'/'+str(today.day)
date = datetime.datetime.strptime(str1, "%y/%m/%d")
date = date + datetime.timedelta(days=-1)
date = str(date.year)+str(date.month)+str(date.day)
```

```python
def check_in_eng(word):
    sent = stemmer.stem(word)
    res = " ".join(w for w in nltk.wordpunct_tokenize(sent) if w.lower() in words or not w.isalpha())
    return res
```

```python
import string
def isEnglish(s):
      return s.isascii()
```

```python
path = date+'_post_btc.csv'

df = pd.read_csv(path)
#date = path[0:4]+'/'+path[4:6]+'/'+path[6:8]
#date = path[0:4]+'/'+path[4:6]+'/'+path[6:8]
key_word = []
for i in range(df.shape[0]):
    k = df['key_word_btc'][i][1:len(df['key_word_btc'][i])-1].split(',')
    for j in range (len(k)):
        k[j] = k[j].strip()
        k[j] = k[j][1:len(k[j])-1]
        lis = k[j].split()
        for i in range(len(lis)):
            lis[i] = check_in_eng(lis[i])
        res = " ".join(lis)
        key_word.append(res)
        key_word.append(k[j])
dff = pd.DataFrame(key_word)
k = pd.DataFrame(dff.apply(pd.value_counts))
top_keybtc = []
remove = ['bitcoin','btc','crypto']
i=0
while len(top_keybtc) < 5:
    cur = k.index[i].lower()
    cur = cur.strip()
    if cur and cur not in remove:
        top_keybtc.append(cur)
    i = i+1
score_btc = df['score_btc'][0]
#sentiment_btc = df["Sentiment"].value_counts().index[0]
if score_btc > 5:
    sentiment_btc = 'Positive'
elif score_btc < 5:
    sentiment_btc = 'Negative'
else:
    sentiment_btc = 'Neutral'
```

```python
try:
    to_unicode = unicode
except NameError:
    to_unicode = str

data = {
    'sentiment_result':{
        'twitter':{
            'bitcoin':
            [
             {
                 'date':str(date),
              'sentiment': sentiment_btc,
              'score':score_btc,
              'keywords':top_keybtc
             }
            ]
        }
    }
}
```

```python
with io.open(date+'.json', 'w', encoding='utf8') as outfile:
    str_ = json.dumps(data,
                      indent=4, sort_keys=True,
                      separators=(',', ': '), ensure_ascii=False)
    outfile.write(to_unicode(str_))
```
